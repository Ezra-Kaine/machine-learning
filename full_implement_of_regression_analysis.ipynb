```python
from google.colab import files
uploaded = files.upload()

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('xydataset (1).csv')
df.head()

X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values

X = (X - X.mean(axis=0)) / X.std(axis=0)

X = np.c_[np.ones(len(X)), X]

theta = np.zeros(X.shape[1])
learning_rate = 0.01
epochs = 1000
n = len(y)

for i in range(epochs):
    y_pred = X.dot(theta)
    error = y_pred - y

    gradients = (1/n) * X.T.dot(error)
    theta = theta - learning_rate * gradients

print("Model parameters (theta):", theta)
print("Regression Equation:")
print("z = theta0 + theta1*x + theta2*y")

plt.scatter(y, X.dot(theta))
plt.xlabel("Actual Z values")
plt.ylabel("Predicted Z values")
plt.title("XYZ Regression using Gradient Descent")
plt.show()
```
