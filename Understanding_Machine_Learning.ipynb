```python
import nltk
from nltk.chat.util import Chat

my_reflections = {
    'i am':'you are',
    'i was':'you were',
    'i':'you',
    'im':'you are',
    'id':'you would',
    'ive':'you have',
    'ill':'you will',
    'my':'your',
    'you are':'i am',
    'you were':'i was',
    'youve':'i have',
    'youll':'i will',
    'your':'my',
    'yours':'mine',
    'you':'me',
    'me':'you'
}

pairs = [
    [r'my name is (.*)', ['Hello %1, how are you today?']],
    [r'hi|hey|hello', ['Hello!', 'Hey there!']],
    [r'what.*your name', ['I am a bot created by Ezra. My name is SteveAI.']],
    [r'how are you', ['I am doing good. How about you?']],
    [r'sorry (.*)', ['No problem.', 'Itâ€™s okay.']],
    [r'im (.*) doing good', ['Nice to hear! What can I do for you today?']],
    [r'.*age.*', ['AI does not have an age.']],
    [r'.*health.*', ['AIs do not need health.']],
    [r'what.*favourite sport', ['Field hockey.']],
    [r'.*weather.*', ['Look outside ']],
    [r'quit', ['Bye! Take care.']]
]

def mychatbot():
    print("Hi, I'm a chatbot created by Ezra. Type 'quit' to exit.")
    chatbot = Chat(pairs, my_reflections)
    chatbot.converse()

if __name__ == '__main__':
    mychatbot()
```
